```python
# https://www.census.gov/programs-surveys/cps/technical-documentation/methodology/industry-and-occupation-classification.html

# https://usa.ipums.org/usa-action/variables/OCCSOC#description_section

import os
import glob

wages_values_2004_2009 = []
wages_values_2010_2017 = []
wages_values_2018_2021 = []

a= 0 
os.chdir(main_link + "Input/ACS PUMS Wages State Level/")
for FileList in glob.glob('*.csv'):
        Year = FileList[15:19]
        ACS_PUMS_wage_df = pd.read_csv(main_link + "Input/ACS PUMS Wages State Level/" +FileList )
        wages_long = ACS_PUMS_wage_df.melt(id_vars=['SOC occuation code 5%'], value_vars=ACS_PUMS_wage_df.columns.tolist()[1:])
        wages_long["Year"] = Year
        wages_long["Occ_Group"] = wages_long['SOC occuation code 5%'].apply(lambda x : x[:3].lower())
        wages_long["SOC_code"] = wages_long['SOC occuation code 5%'].apply(lambda x : re.sub(r'[^\w]', ' ', x[4:]).lower().strip())
        wages_long_values = wages_long[["Occ_Group","SOC_code", "Year", "variable", "value" ]].values.tolist()  #
        wages_long_values = [x for x in wages_long_values if x[0].isnumeric() == False]
        
        if int(Year)<2010: wages_values_2004_2009.extend(wages_long_values)
        elif int(Year)<2018: wages_values_2010_2017.extend(wages_long_values)
        elif int(Year)>2017: wages_values_2018_2021.extend(wages_long_values)
        a= a+len(wages_long_values )

print(a)
print(len(wages_values_2004_2009))
print(len(wages_values_2010_2017))
print(len(wages_values_2018_2021))

# occ_industry_pre = []
# for i in wages_values_all: occ_industry_pre.append([i[1], i[0]])



# occ_industry = []
# for l in occ_industry_pre: 
#   if l not in occ_industry : occ_industry.append(l)

# print(len(occ_industry_pre))
# print(len(occ_industry))
```
